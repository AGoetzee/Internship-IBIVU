```python
import plotly.graph_objects as go
import numpy as np
```

```python
xs = np.arange(0.0,3.0,0.1)
ys = np.arange(0.0,3.0,0.1)
```

```python
def calculate_z(xs,ys,function='addition'):
    z = np.zeros((xs.shape[0],ys.shape[0]))
    if function=='addition':
        for i,x in enumerate(xs):
            for j,y in enumerate(ys):
                z[i][j] = x + y
    if function=='subtraction':
        for i,x in enumerate(xs):
            for j,y in enumerate(ys):
                z[i][j] = x - y
    if function=='division':
        for i,x in enumerate(xs):
            for j,y in enumerate(ys):
                z[i][j] = x/y
    if function=='multiplication':
        for i,x in enumerate(xs):
            for j,y in enumerate(ys):
                z[i][j] = x * y
    return z

def plot_z_surface(x,y,z,name):
    fig = go.Figure(data=[go.Surface(z=z, x=xs, y=ys)])

    fig.update_layout(title=name, autosize=False,
                      width=500, height=500,
                      margin=dict(l=65, r=50, b=65, t=90))

    fig.show()
```

```python
for scheme in ['addition','subtraction','division','multiplication']:
    z = calculate_z(xs,ys,scheme)
    plot_z_surface(xs,ys,z,scheme)
```
