```python
plt.clf()
plt.figure(dpi=150)
p=plt.plot(time,dist.mean(axis=0),alpha=1,label = 'Mean separation')
plt.errorbar(time,dist.mean(axis=0),yerr=dist.std(axis=0),alpha=0.01,color=p[0].get_color())
plt.xlabel('Time (ps)')
plt.ylabel('COM separation (nm)')
plt.title('COM separation over time (n=5)')
plt.legend()
```

```python
plt.clf()
plt.figure(dpi=150)
p=plt.plot(time,bias.mean(axis=0),alpha=1,label = 'Mean bias')
plt.errorbar(time,bias.mean(axis=0),yerr=bias.std(axis=0),alpha=0.01,color=p[0].get_color())
plt.xlabel('Time (ps)')
plt.ylabel('Bias potential (kj/mol)')
plt.title('Metadynamics bias over time (n=5)')
plt.legend()
```
