```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.clf()
plt.figure(dpi=150)
for i in range(0,150,1):
    data = np.loadtxt(f'fes_{i}.dat').T
    plt.plot(data[0],data[1],label=str(i))
plt.xlim(-1,10)
plt.xlabel('COM separation (nm)')
plt.ylabel('Relative free energy (kj/mol)')
plt.title('Free energy through time (stride=100, 0.2 ps)')
```
