```python
import numpy as np
import pandas as pd

def colvar_to_df(f):
    colnames=f.readline().strip('#! FIELDS ').split()
    return pd.read_table(f,skiprows=0,header=None,sep=' ',names=colnames).set_index(colnames[0])
```

```python
with open('par-restr-rmsd-2\COLVAR_par-restr-rmsd-2','r') as f:
    df = colvar_to_df(f)
df.head()
```

```python
df[['c1','c27']].plot.hist()
```

```python
df[['c1','c27','t1']].plot(alpha=0.5)
```

```python
df['t1'].plot.hist(bins=50)
```

```python
df['restraint.bias'].plot()
```

```python
df[['combine','c1','c27']].plot(alpha=0.5)
```
