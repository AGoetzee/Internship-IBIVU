```python
import pandas as pd
import numpy as np
import plumed
import os
import matplotlib.pyplot as plt
```

```python
folders = ['noSASA', 'SASA']
temps = [280,300,320]
GRID_MIN = -0.5
GRID_MAX = 9.5
BINS = 500
KT = {280: 2.328049,
      300: 2.494339,
      320: 2.660628}
```

```python
def get_fes(infile,outfile,kt):
    os.system(f'plumed sum_hills --hills {infile} --outfile {outfile} --min {GRID_MIN} --max {GRID_MAX} --bin {BINS} --kt {kt} --mintozero ')
    return

def get_convergence(infile,outfile,kt,stride=2000):
    os.system(f'plumed sum_hills --hills {infile} --outfile {outfile} --stride {stride} --min {GRID_MIN} --max {GRID_MAX} --bin {BINS} --kt {kt} --mintozero ')
```

```python
for folder in folders:
    for temp in temps:
        
        # Get fes
        path = f'{folder}/{temp}/HILLS'
        # if not os.path.exists(path):
        #     print(f'{folder}/{temp}/HILLS does not exist')
        #     continue
            
        out = f'{folder}/{temp}/fes.dat'
        # if not os.path.exists(out):
        # get_fes(path,out,KT[temp])
        if not os.path.exists(f'{folder}/{temp}/convergence/'):
            os.mkdir(f'{folder}/{temp}/convergence')
        get_convergence(path,f'{folder}/{temp}/convergence/fes',KT[temp])
        
```

```python
fig = plt.figure(constrained_layout=True, dpi=300)

fig.suptitle('Metadynamics - COM Separation\nRelative Free Energies')
fig.supylabel('G kJ/mol')
fig.supxlabel('COM Separation (nm)')

subfigs = fig.subfigures(nrows=2, ncols=1)

for i, folder in enumerate(folders):
    subfigs[i].suptitle(f'{folder} model')
    axis = subfigs[i].subplots(nrows=1, ncols=3)
    for j, temp in enumerate(temps):
            
        out = f'{folder}/{temp}/fes.dat'
        
        fes = plumed.read_as_pandas(out)
        axis[j].plot(fes['dist'],fes['file.free'], label = f'{folder} - {temp}')
        axis[j].set_xlim([-0,6.9])
        axis[j].set_title(f'{temp} K')

        del fes
        

# plt.legend()   
```

```python
fig, axis = plt.subplots(2, 1, constrained_layout=True, dpi=300)

fig.suptitle('Metadynamics - COM Separation\nRelative Free Energies')
fig.supylabel('G kJ/mol')
fig.supxlabel('COM Separation (nm)')

for i, folder in enumerate(folders):
    axis[i].set_title(f'{folder} model')
    for j, temp in enumerate(temps):
            
        out = f'{folder}/{temp}/fes.dat'
        fes = plumed.read_as_pandas(out)
        
        axis[i].plot(fes['dist'],fes['file.free'], label = f'{temp} K')
        axis[i].set_xlim([-0,6.9])

        del fes
axis[i].legend(loc='upper right')   
```

```python
fig, (axis, axis2) = plt.subplots(1,2,constrained_layout=True, dpi=600)
folder = 'SASA'
temp = 300
times = [4,14,24,29,34,49]

# Left
fes = plumed.read_as_pandas(f'{folder}/{temp}/fes.dat')
axis.plot(fes['dist'],fes['file.free'])
axis.set_xlim([-0,6.9])
axis.set_title('Free Energy')


# Right
for i in times:
            
    out = f'{folder}/{temp}/convergence/fes{i}.dat'
    fes = plumed.read_as_pandas(out)

    axis2.plot(fes['dist'],fes['file.free'], label = f'{i+1} ns')
    axis2.set_xlim([-0,6.9])

    del fes
axis2.set_title('Convergence')
    
# Rest of the fig
fig.legend(bbox_to_anchor=(1.0, 0.83),loc='upper left', borderaxespad=0.)
fig.supxlabel('COM Separation (nm)')
fig.supylabel('Relative Free Energy (kJ/mol)')
fig.suptitle(f'Metadynamics \n {temp} K {folder} model')
plt.savefig('MetadExample.png',bbox_inches='tight')
```
